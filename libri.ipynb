```python
import numpy as np
import torch
import os
from os import listdir
from os.path import isfile, isdir, join
import ntpath
import librosa
import hgtk
import soundfile as sf
from scipy.io.wavfile import read

def find_wav(path, data_files):
    for file in listdir(path):
        if isfile(join(path, file)) and 'wav' in file:
            data_files.append(join(path, file))
        elif isdir(join(path, file)):
            data_files = find_wav(path + '/' + file, data_files)
        
    return data_files

root_dir = '/home/scpark/ai/datasets/LibriTTS/train-clean-360'

data_files = []
data_files = find_wav(root_dir, data_files)
data_files.sort()
print(len(data_files))
```

```python
train_ratio, valid_ratio, test_ratio = 0.95, 0.99, 1
train_index = int(len(data_files) * train_ratio)
valid_index = int(len(data_files) * valid_ratio)
test_index = int(len(data_files) * test_ratio)
print(train_index, valid_index, test_index)

def write_filelist(filename, data_files):
    from tqdm.notebook import tqdm
    
    with open(filename, 'w') as file:

        for index, data_file in tqdm(enumerate(data_files)):
            
            sampling_rate, data = read(data_file)
            if len(data) > sampling_rate * 10:
                print(len(data))
                continue
                
            txt_file = data_file[:-4] + '.normalized.txt'

            with open(txt_file) as f:
                line = f.readline()

            l = data_file + '|' + line
                
            file.write(l + os.linesep)

write_filelist('filelists/libri_audio_text_train_filelist.txt', data_files[:train_index])
write_filelist('filelists/libri_audio_text_val_filelist.txt', data_files[train_index:valid_index])
write_filelist('filelists/libri_audio_text_test_filelist.txt', data_files[valid_index:test_index])
```
